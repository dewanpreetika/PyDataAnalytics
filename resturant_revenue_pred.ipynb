```python
# setting the feature variable (x)
x = updated_dataset.drop('Revenue', axis=1)
x.head(2)
```

```python
y = dataset['Revenue']
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test, y_train, y_test = train_test_split(x , y, test_size=0.4)

## checking the dimensions of the training and testing data
print(f"x_train:{x_train.shape} , x_test:{x_test.shape} y_train:{y_train.shape} , y_test:{y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
# fitting the model
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
mse = mean_squared_error(y_test, y_pred)
r2Score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Squared Error: ",mse)
print("R squared Error: ",r2Score)
print("Mean Absolute Error: ",mae)
print("R2 score for training: ",  model.score(x_train,y_train))
print("R2 score for testing", model.score(x_test, y_test))
```
